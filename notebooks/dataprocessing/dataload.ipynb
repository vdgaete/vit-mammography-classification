```python
# Notebook para converter arquivos dicom
# 1 Passo
import os
import pydicom
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2


# Carregar arquivos do CBIS-DDSM
# O dataset CBIS-DDSM é um dataset de mamografias digitais, que contém imagens de mamografias digitais

DATASET_TRAIN = 'D:\Data\CBISDATASET\Train'
DATASET_TEST = 'D:\Data\CBISDATASET\Test'

# Carregar descrição e metadados do dataset
mass_train_description = pd.read_csv(DATASET_TRAIN+'\mass_case_description_train_set.csv')
mass_test_description = pd.read_csv(DATASET_TEST+'\mass_case_description_test_set.csv')


# Data Exploration

# Verificar a descrição do dataset
print(mass_train_description.head())
print(mass_test_description.head())




# Cropped image view
# Carregar uma imagem do dataset
originalpath = mass_train_description['cropped image file path'][0]
# Change the file path directory separator
originalpath = originalpath.replace("/", "\\")
# Remove filename
file_p = originalpath.split("\\")
wrongname = file_p[-1]
filedir = originalpath.replace(wrongname, "")
# Verificar o arquivo menor 

filename = ''
last = 9999999999999999
for file in os.listdir(DATASET_TRAIN + '\\CBIS-DDSM\\' + filedir):
        # check  filesize
    if file.endswith('.dcm'): 
        size = os.path.getsize(DATASET_TRAIN + '\\CBIS-DDSM\\' + filedir + file)
        print(file, size)
        if size < last:
            last = size
            filename = file



filepath = filedir + filename

img = pydicom.dcmread(DATASET_TRAIN + '\\CBIS-DDSM\\' + filepath)

# Visualizar a imagem
plt.imshow(img.pixel_array, cmap='gray')
plt.show()


# Save the image as a png file
cv2.imwrite('D:\Data\CBISDATASET\MASSTRAIN\cropped_image.png', img.pixel_array)

## Label Cleaning for Mass Train Description

# Only CC View (Remove MLO View)
mass_train_description = mass_train_description[mass_train_description['image view'] == 'CC']

# Remove ROI mask column
mass_train_description = mass_train_description.drop(columns=['ROI mask file path'])

# Remove Image view column
mass_train_description = mass_train_description.drop(columns=['image view'])

# Remove Image Path column
mass_train_description = mass_train_description.drop(columns=['image file path'])

# Remove Abnormality Type column
mass_train_description = mass_train_description.drop(columns=['abnormality type'])

# Remove Abnormality ID column
mass_train_description = mass_train_description.drop(columns=['abnormality id'])

# Rename Cropped Image Path column
mass_train_description = mass_train_description.rename(columns={'cropped image file path': 'image file path'})

# Remove Mass Shape column
mass_train_description = mass_train_description.drop(columns=['mass shape'])

# Remove Mass Margin column
mass_train_description = mass_train_description.drop(columns=['mass margins'])

# Remove Assessment column
mass_train_description = mass_train_description.drop(columns=['assessment'])

# Remove Subtlety column
mass_train_description = mass_train_description.drop(columns=['subtlety'])


# check the small file size
for index, row in mass_train_description.iterrows():
    originalpath = row['image file path']
    # Change the file path directory separator
    originalpath = originalpath.replace("/", "\\")
    # Remove filename
    file_p = originalpath.split("\\")
    wrongname = file_p[-1]
    filedir = originalpath.replace(wrongname, "")
    # Verificar o arquivo menor

    filename = ''
    last = 999999999999999999999
    for file in os.listdir(DATASET_TRAIN + '\\CBIS-DDSM\\' + filedir):
        # check  filesize
        size = os.path.getsize(DATASET_TRAIN + '\\CBIS-DDSM\\' + filedir + file)
        if size < last:
            last = size
            filename = file

    filepath = filedir + filename
    mass_train_description.at[index, 'image file path'] = filepath

    

## Label Cleaning for Mass Test Description

# Only CC View (Remove MLO View)
mass_test_description = mass_test_description[mass_test_description['image view'] == 'CC']

# Remove ROI mask column
mass_test_description = mass_test_description.drop(columns=['ROI mask file path'])

# Remove Image view column
mass_test_description = mass_test_description.drop(columns=['image view'])

# Remove Image Path column
mass_test_description = mass_test_description.drop(columns=['image file path'])

# Remove Abnormality Type column
mass_test_description = mass_test_description.drop(columns=['abnormality type'])

# Remove Abnormality ID column
mass_test_description = mass_test_description.drop(columns=['abnormality id'])

# Rename Cropped Image Path column
mass_test_description = mass_test_description.rename(columns={'cropped image file path': 'image file path'})

# Remove Mass Shape column
mass_test_description = mass_test_description.drop(columns=['mass shape'])

# Remove Mass Margin column
mass_test_description = mass_test_description.drop(columns=['mass margins'])

# Remove Assessment column
mass_test_description = mass_test_description.drop(columns=['assessment'])

# Remove Subtlety column
mass_test_description = mass_test_description.drop(columns=['subtlety'])

# check the small file size
for index, row in mass_test_description.iterrows():
    originalpath = row['image file path']
    # Change the file path directory separator
    originalpath = originalpath.replace("/", "\\")
    # Remove filename
    file_p = originalpath.split("\\")
    wrongname = file_p[-1]
    filedir = originalpath.replace(wrongname, "")
    # Verificar o arquivo menor

    filename = ''
    last = 99999999999999999999
    for file in os.listdir(DATASET_TEST + '\\CBIS-DDSM\\' + filedir):
        # check  filesize
        size = os.path.getsize(DATASET_TEST + '\\CBIS-DDSM\\' + filedir + file)
        if size < last:
            last = size
            filename = file

    filepath = filedir + filename
    mass_test_description.at[index, 'image file path'] = filepath


# Save the cleaned data

mass_train_description.to_csv(DATASET_TRAIN+'\\clean_mass_train_description.csv', index=False)
mass_test_description.to_csv(DATASET_TEST+'\\clean_mass_test_description.csv', index=False)

# Header
print(mass_train_description.head())
print(mass_test_description.head())


```
