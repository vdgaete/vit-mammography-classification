```python
# 2 Passo
import os
import pydicom
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2


# Carregar arquivos do CBIS-DDSM
# O dataset CBIS-DDSM é um dataset de mamografias digitais, que contém imagens de mamografias digitais

DATASET_TRAIN = 'D:\Data\CBISDATASET\Train'
DATASET_TEST = 'D:\Data\CBISDATASET\Test'


# Carregar descrição e metadados do dataset
mass_train_description = pd.read_csv(DATASET_TRAIN+'\clean_mass_train_description.csv')
mass_test_description = pd.read_csv(DATASET_TEST+'\clean_mass_test_description.csv')

clean_mass_train_description = mass_train_description
clean_mass_test_description = mass_test_description

# Verificar a descrição do dataset
print(mass_train_description.head())
print(mass_test_description.head())

# Converter dcm para png
def dcm2png(dcm_path, png_path):
    ds = pydicom.dcmread(dcm_path)
    img = ds.pixel_array
    cv2.imwrite(png_path, img)

#Mkdir
if not os.path.exists(DATASET_TRAIN + '\\Train'):
    os.makedirs(DATASET_TRAIN + '\\Train')
if not os.path.exists(DATASET_TEST + '\\Test'):
    os.makedirs(DATASET_TEST + '\\Test')


# Converter todas as imagens de treino de dcm para png
print('Converting train images...')
for index, row in mass_train_description.iterrows():
    print('Converting image ' + str(index) + ' of ' + str(len(mass_train_description)))
    # New name of the image
    OldName = row['image file path'].split('\\')[-1]
    NewName = row['image file path'].split('\\')[0] + '.png'

    print('Old Name: ' + OldName)
    print('New Name: ' + NewName)
    newPath = DATASET_TRAIN + '\\Train\\' + NewName
    # Save File Path
    clean_mass_train_description.at[index, 'image file path'] = NewName
    dcm2png(DATASET_TRAIN + '\\CBIS-DDSM\\' + row['image file path'], newPath)

    
# Converter todas as imagens de teste de dcm para png
print('Converting test images...')
for index, row in mass_test_description.iterrows():
    print('Converting image ' + str(index) + ' of ' + str(len(mass_test_description)))
    # New name of the image
    OldName = row['image file path'].split('\\')[-1]
    NewName = row['image file path'].split('\\')[0] + '.png'

    print('Old Name: ' + OldName)
    print('New Name: ' + NewName)
    newPath = DATASET_TEST + '\\Test\\' + NewName
    # Save File Path
    clean_mass_test_description.at[index, 'image file path'] = NewName
    dcm2png(DATASET_TEST + '\\CBIS-DDSM\\' + row['image file path'], newPath)
```

```python
# Salvar descrição do dataset
clean_mass_train_description.to_csv(DATASET_TRAIN + '\\clean_mass_train_description2.csv', index=False)
clean_mass_test_description.to_csv(DATASET_TEST + '\\clean_mass_test_description2.csv', index=False)
```
