```python
# Plot history results
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
import json
import pandas as pd

def plot_histories(histories):
    for history in histories:
        plt.figure(figsize=(10, 5))
        plt.plot(history['loss'], label=history['name'][0] + ' loss')
        plt.plot(history['val_loss'], label=history['name'][0] + ' val_loss')
        plt.title('model loss')
        plt.ylabel('loss')
        plt.xlabel('epoch')
        plt.legend(loc='upper right')
        plt.show()

def plot_histories_acc(histories):
    
    for history in histories:
        plt.figure(figsize=(10, 5))
        plt.plot(history['accuracy'], label=history['name'][0] + ' acc')
        plt.plot(history['val_accuracy'], label=history['name'][0] + ' val_acc')
   
        plt.title('model accuracy')
        plt.ylabel('accuracy')
        plt.xlabel('epoch')
        plt.legend(loc='upper right')
        plt.show()
        
# Load history csv files
histories = []
current_path = os.getcwd()
for file in os.listdir(current_path):
    if file.endswith(".csv"):
        history = pd.read_csv(file)
        history['name'] = file
        print(file)
        histories.append(history)




plot_histories(histories)
plot_histories_acc(histories)


```
